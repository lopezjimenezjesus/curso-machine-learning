> **Ejercicio**: Defina una función que convierta grados centígrados en grados Farenheit.

```python
#<SOL>

def celsiusToFarenheit(celsius):
  return (celsius * 9 / 5) + 32

print(celsiusToFarenheit(32))

#</SOL>
```

> **Ejercicio**: Diseñe una función que devuelva la solución de la ecuación lineal ax+b=0, dados a y b. Si la ecuación tiene infinitas soluciones o no tiene solución, la función lo detectará y devolverá None.

```python
#<SOL>

def ecuacion(a,b):
  if a == 0:
    return None
  x = -b / a 
  return x

print(ecuacion(1,2))
print(ecuacion(0,1))


#</SOL>
```

> **Ejercicio**: Diseñe un programa que dado un número entero, muestre por pantalla sus cifras en orden inverso. Cada cifra aparecerá en una línea diferente. Ojo, aquí no devolvemos ningún valor.

```python
#<SOL>

def inverso(n):
  l = len(str(n))
  for x in range(l-1,-1,-1):
    print(str(n)[x])
    
inverso(1235)
#</SOL>
```

> **Ejercicio EXTRA**: Diseñe una función que nos diga si un número dado como parámetro es perfecto o no. Un número es perfecto si es igual a la suma de todos sus divisores excepto él mismo.




```python
# <SOL>
def perfecto(n):
  suma = 0
  for x in range(1,n):
    if n % x == 0:
      suma += x
  if suma == n:
    print(n, "es perfecto")
  else:
    print(n, "no es perfecto")
  
perfecto(3)
perfecto(8)
perfecto(496)
# </SOL>
```

> **Ejercicio EXTRA!**: Diseñe un programa que devuelva los números perfecto comprendidos entre 1 y N, siendo N un número pasado como argumento.

```python
#<SOL>

def n_perfectos(n):
  for i in range(1, n+1):
    print("Número:", i)
    suma = 0
    for j in range(1,i):
      if i % j == 0:
        suma += j
    if suma == i:
      print(i, "es perfecto \n")
    else:
      print(i, "no es perfecto \n")
print(n_perfectos(497))
#</SOL>
```
