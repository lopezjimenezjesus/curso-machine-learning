> **Ejercicio**: Diseñe un programa que, dados 5 números enteros, determine cuál de los 4 últimos números es más cercano al primero.

```python
#<SOL>

n_1 =  int(input("Introduce un numero entero"))
n_2 =  int(input("Introduce otro numero entero"))

dist = abs(n_1 - n_2)
mas_cercano = n_2

for i in range(3):
  n  =  int(input("Introduce otro numero entero"))
  if abs(n_1 - n) < dist:
    mas_cercano  = n
  elif abs(n_1 - n) == dist:
    mas_cercano = str(n) + ',' + str(mas_cercano)

print(mas_cercano)


#</SOL>
```

> **Ejercicio**: Diseñe un programa para calcular si un número es primo o no.

```python
#<SOL>

num = int(input("Introduce un número: "))
counter = 0

for n in range(2, num):
  if num % n == 0:
      print(n, "es divisor")
      counter += 1

if counter == 0:
  print("Es primo")


#</SOL>
```
